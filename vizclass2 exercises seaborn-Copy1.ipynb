```python
# import the libraries including seaborn and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Set the parameters
# Take some time to play around with the settings you like and try them out on the basic graph below before settling

# set the style
sns.set_style('darkgrid')
# set the palette - generate a palplot to show what you will use
sns.set_palette('tab10',10)
# set the context
sns.set_context('talk')
```

```python
# genearte a palplot of your chosen palatte
colors=['pig pink','deep aqua', 'burple']
sns.palplot(sns.xkcd_palette(colors))
```

```python
# Use this to try out your settings
x = np.linspace(1,20, 2)
sns.lineplot(x, x)
sns.lineplot(x, x + 1)
sns.lineplot(x, x + 2)
sns.lineplot(x, x + 3)
sns.lineplot(x, x + 4)
sns.lineplot(x, x + 5)
```

```python
# read in the tips dataset
tips= sns.load_dataset('tips')
tips.head
```

```python
# plot a simple histogram of the tips paid
# remove the kde line, show only 10 bins
sns.distplot(tips['tip'],kde= False, bins=10 )
```

```python
# Generate a line plot showing both total bill and tips on the same axis, using index number as the independent variable
# Hine you will need to use sns.lineplot twice

sns.lineplot(tips.index, tips.total_bill)
sns.lineplot(tips.index, tips.tip)
```

```python
# Building on the above add a suitable title, axis labels and make it wider using matplotlib features
fig, ax = plt.subplots(figsize=(20,6))
sns.lineplot(tips.index, tips.total_bill)
sns.lineplot(tips.index, tips.tip)
plt.title('A graph showing the total bill and tips by index number')
plt.xlabel('Index number')
plt.ylabel('Total bill and tips')

```

```python
# Generate a lineplot again but this time show the tipped amount split by sex
# Hint you will need to change the format of the parameters and use the hue feature
```

```python
#sns.lineplot(tips.index, tips.total_bill)
sns.lineplot(tips.index, tips.tip, hue= tips.sex)
```

```python
# Create a scatterplot of total bill and amount tipped with correctly labelled axes and title
# You will need to include a regression line so make sure you use the correct kind of plot
fig, ax = plt.subplots(figsize= (10,4))
sns.regplot(tips.total_bill, tips.tip)

```

```python
# Create a second scatterplot without the regression line
# Use the hue feature to create different coloured markers for the different days of the week
```

```python
fig, ax = plt.subplots(figsize= (10,4))
sns.scatterplot(tips.total_bill, tips.tip, hue=tips.day)
```

```python
# Use countplot to show the different numbers of customers on different days of the week
sns.countplot(tips.day)
```

```python
# Use a countplot to show number of customers by different days of the week and split to create new bars showing lunch and dinner
# hint - you will need to use hue again
```

```python
sns.countplot(tips.day, hue= tips.time)
```

```python
# generate a fully labelled boxplot showing the amount males tips vs female tips
sns.boxplot(tips.sex, tips.tip)
```

```python
# modify the above plot to create a violin plot showing the difference in the distrtibution of amount tipped 
# between smoking and non smoking males and smoking and non smoking females
sns.violinplot(tips.sex,tips.tip, hue= tips.smoker)
```

```python
# Use lm plot to create a grid showing the amount tipped by males and females according to different days of the week
# Use the documentation to create neat layout and clear and correct labels and titles
sns.lmplot(x='total_bill', y='tip', row='sex', col='day', data= tips, height=3, aspect=2)

```
