## simple single plot

```python
### import pandas, numpy and matplotlib libraries according to their conventions
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### read in the 'key_data.csv' file as 'data'
data= pd.read_csv('key_data.csv')
data.tail()
```

```python
data['year'].max()
```

```python
data['year'].min()
```

```python
# Using matplotlib generate a simple plot showing Ugandas GDP per capita over time and China's gdp 
# on the same axes over time using the gdp_ug and gdp_ch column

plt.plot(data.year,data.gdp_ug)
plt.plot(data.year,data.gdp_ch)
plt.show()
```

```python
# Build on the above plot by adding a suitable x axis label, y axis label and title

plt.plot(data.year,data.gdp_ug)
plt.plot(data.year,data.gdp_ch)
plt.xlabel('Year')
plt.ylabel('Gross Domestic Product')
plt.title("A Graph showing Uganda's GDP against China's")
plt.show()

```

```python
# Build on the above plot by changing the line colour, style and width of the lines
plt.plot(data.year,data.gdp_ug, color='Green', lw=2, ls= '-')
plt.plot(data.year,data.gdp_ch, color='Red', lw=2, ls= '-')
plt.xlabel('Year')
plt.ylabel('Gross Domestic Product')
plt.title("A Graph showing Uganda's GDP against China's")
plt.show()

```

```python
# Build on the above plot by adding markers, they are always ugly so make them as ugly as you like
plt.plot(data.year,data.gdp_ug, color='Green', lw=2, ls= '-', marker='+')
plt.plot(data.year,data.gdp_ch, color='Red', lw=2, ls= '-', marker='o')
plt.xlabel('Year')
plt.ylabel('Gross Domestic Product')
plt.title("A Graph showing Uganda's GDP against China's")
plt.show()
```

```python
# BUild on the above plot by adding a legend anwywher you think is good
plt.plot(data.year,data.gdp_ug, color='Green', lw=2, ls= '-', marker='+', label='Uganda')
plt.plot(data.year,data.gdp_ch, color='Red', lw=2, ls= '-', marker='o',label='China' )
plt.xlabel('Year')
plt.ylabel('Gross Domestic Product')
plt.title("A Graph showing Uganda's GDP against China's")
plt.legend()
plt.show()
```

# Multiple plots
Remember for this you must use the fig, ax = plt.subplots() format

```python
# Create a figure with 3 X  1 or 1 X 3 subplots showing gdp per capita between 1980 and 2020
# Each plot should have its own colour and style
# Use judgement to decide wehther to change the ticks, use log scale etc
# subplots should have no top or side spines
# x axis ticks should be at 5 year intervals
# be sure to include a suptitle for the whole image

fig,(ax1,ax2,ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10,20), dpi=100)
#setting parameters for Uganda Graph
ax1.plot(data.year, data.gdp_ug, color='Green', label='Uganda', lw=2, ls=':')
ax1.set_xlabel('year')
ax1.set_ylabel('Gross Domestic Product')
ax1.set_title('Gross Domestic Product of Uganda')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.set_xticks(np.arange(1985,2020,5))
ax1.set_yscale('log')

#setting parameters for China Graph
ax2.plot(data.year, data.gdp_ch, color='Red', label='China', lw=2, ls='-')
ax2.set_xlabel('year')
ax2.set_ylabel('Gross Domestic Product')
ax2.set_title('Gross Domestic Product of China')
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.set_xticks(np.arange(1960,2020,5))
ax2.ticklabel_format(style='sci' ,axis='y' ,scilimits=(5,5) )

#setting parameters for France Graph
ax3.plot(data.year, data.gdp_ch, color='Blue', label='France', lw=2, ls='-.')
ax3.set_xlabel('year')
ax3.set_ylabel('Gross Domestic Product')
ax3.set_title('Gross Domestic Product of France')
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax3.set_xticks(np.arange(1960,2020,5))

#setting parameters for the whole figure
plt.suptitle('A figure showing graphs of GDP of Uganda, China and France overtime', y=1.05)
plt.tight_layout()
plt.show()
```
